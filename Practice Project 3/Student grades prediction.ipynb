```python
import pandas as pd
```

```python
# Load the dataset
url = 'https://raw.githubusercontent.com/dsrscientist/dataset4/main/Grades.csv'
df = pd.read_csv(url)
```

```python
# Explore the dataset
print(df.head()) # Print the first 5 rows of the dataset
print(df.info()) # Print information about the dataset, such as the data types and missing values
print(df.describe()) # Print descriptive statistics for the numerical attributes
```

```python
# Visualize the dataset
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.pairplot(df, vars=['CGPA', 'AB-101', 'AB-201', 'AB-301']) # Plot pairwise relationships between the target variable and some of the course grades
plt.show()
```

```python
sns.histplot(df, x='CGPA', bins=10) # Plot a histogram of the target variable
plt.show()
```
