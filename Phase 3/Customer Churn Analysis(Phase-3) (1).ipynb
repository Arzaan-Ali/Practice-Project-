```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



```

```python
# Load the dataset from the provided link
url = 'https://github.com/dsrscientist/DSData/raw/master/Telecom_customer_churn.csv'
df = pd.read_csv(url)
```

```python
# Count the number of churned and non-churned customers
churn_counts = df['Churn'].value_counts()
```

```python
# Create a bar graph
plt.figure(figsize=(8, 6))
sns.barplot(x=churn_counts.index, y=churn_counts.values)
```

```python
plt.title('Churned vs Non-Churned Customers')
plt.xlabel('Churn')
plt.ylabel('Count')
plt.show()
```
