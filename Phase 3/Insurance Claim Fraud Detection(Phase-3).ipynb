```python
import pandas as pd

# Load the dataset
url = 'https://raw.githubusercontent.com/dsrscientist/Data-Science-ML-Capstone-Projects/master/Automobile_insurance_fraud.csv'
data = pd.read_csv(url)

# Display the first few rows of the dataset
print(data.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


```

```python
# Count the number of fraudulent and non-fraudulent claims
fraud_counts = data['fraud_reported'].value_counts()
```

```python
# Create a bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x=fraud_counts.index, y=fraud_counts.values)
```

```python
plt.xlabel('Fraudulent Claim')
plt.ylabel('Count')
```

```python
plt.title('Distribution of Fraudulent vs. Non-Fraudulent Claims')
plt.show()
```
